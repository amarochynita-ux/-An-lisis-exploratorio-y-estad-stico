```python
# Fusiona los datos de llamadas, minutos, mensajes e Internet con base en user_id y month
fusion_data = megaline_calls.merge( megaline_messages, on=['user_id', 'month'], how='outer')
fusion_data = fusion_data.merge(megaline_internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index(), on=['user_id', 'month'], how='outer')

print(fusion_data.head())
```

[Calcula los ingresos mensuales por usuario (resta el l√≠mite del paquete gratuito del n√∫mero total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; a√±ade la tarifa mensual en funci√≥n del plan de llamadas). Nota: Dadas las condiciones del plan, ¬°esto podr√≠a no ser tan trivial como un par de l√≠neas! As√≠ que no pasa nada si dedicas algo de tiempo a ello.]

```python
import pandas as pd
import numpy as np
def calculate_monthly_revenue(row, plans):
    plan_name = row['plan']
    plan_info = plans[plans['plan_name'] == plan_name].iloc[0]


    messages_limit = plan_info['messages_included']
    minutes_limit = plan_info['minutes_included']
    data_limit_gb = plan_info['usd_per_gb']
    monthly_fee = plan_info['usd_monthly_fee']


    extra_message_cost = plan_info['usd_per_message']
    extra_minute_cost = plan_info['usd_per_minute']
    extra_data_cost_gb = plan_info['usd_per_gb']


    actual_messages = row['message_date'] if pd.notna(row['message_date']) else 0
    actual_minutes = row['duration_rounded'] if pd.notna(row['duration_rounded']) else 0
    actual_data_mb = row['mb_used'] if pd.notna(row['mb_used']) else 0
    actual_data_gb = np.ceil(actual_data_mb / 1024) # Round up extra usage to the nearest GB


    extra_messages = max(0, actual_messages - messages_limit)
    extra_minutes = max(0, actual_minutes - minutes_limit)
    extra_data_gb = max(0, actual_data_gb - data_limit_gb)


    extra_messages_cost = extra_messages * extra_message_cost
    extra_minutes_cost = extra_minutes * extra_minute_cost
    extra_data_cost = extra_data_gb * extra_data_cost_gb


    total_revenue = monthly_fee + extra_messages_cost + extra_minutes_cost + extra_data_cost

    return total_revenue


fusion_data['monthly_revenue'] = fusion_data.apply(lambda row: calculate_monthly_revenue(row, megaline_plans), axis=1)

print("\nDataFrame with monthly revenue per user and month:")
print(fusion_data.head())
```

## Prueba las hip√≥tesis estad√≠sticas

[Prueba la hip√≥tesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hip√≥tesis nula y alternativa, escoge la prueba estad√≠stica, determina el valor alfa.]

```python
#Prueba de hip√≥tesis
import numpy as np
from scipy import stats
surf_plan = megaline_plans[megaline_plans['plan_name'] == 'Surf']
ultimate_plan = megaline_plans[megaline_plans['plan_name'] == 'Ultimate']

alpha = 0.05


print("Hip√≥tesis Nula (H0): Los ingresos promedio de los usuarios de los planes Ultimate y Surf son iguales.")
print("Hip√≥tesis Alternativa (H1): Los ingresos promedio de los usuarios de los planes Ultimate y Surf son diferentes.")
print(f"Prueba estad√≠stica seleccionada: Prueba t de Student para dos muestras independientes (asumiendo varianzas desiguales).")
print(f"Nivel de significancia (alfa): {alpha}")


results = st.ttest_ind(surf_plan, ultimate_plan, equal_var=False)

print('Valor p:', results.pvalue)


if (results.pvalue <alpha):
    print("Rechazamos la hip√≥tesis nula.")
    print("Conclusi√≥n: Existe suficiente evidencia estad√≠stica para concluir que los ingresos promedio de los usuarios de los planes Ultimate y Surf son significativamente diferentes.")
else:
    print("No rechazamos la hip√≥tesis nula.")
    print("Conclusi√≥n: No existe suficiente evidencia estad√≠stica para concluir que los ingresos promedios usuarios de los planes Ultimate y Surf son significativamente diferentes.")
```
